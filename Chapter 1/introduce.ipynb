```
inaugural_1933_raw = inaugural.raw('1933-Roosevelt.txt')
inaugural_1933_words = inaugural.words('1933-Roosevelt.txt')
inaugural_1933_sents = inaugural.sents('1933-Roosevelt.txt')

print("First 200 characters of raw text:")
print(inaugural_1933_raw[:200])
print("\nFirst 20 words:")
print(inaugural_1933_words[:20])
print("\nFirst 5 sentences:")
print(inaugural_1933_sents[:5])
```

```
inaugural_1933_raw = inaugural.raw('1933-Roosevelt.txt') # Kode ini mengambil seluruh teks mentah (dalam bentuk string tunggal) dari pidato pelantikan Roosevelt tahun 1933 (dengan ID file '1933-Roosevelt.txt') dari korpus inaugural NLTK dan menyimpannya dalam variabel bernama inaugural_1933_raw.
inaugural_1933_raw # menampilkan isi dari variabel inaugural_1933_raw, yaitu seluruh teks pidato tersebut.
```

#### 2.b.

```
inaugural_1933_words = inaugural.words('1933-Roosevelt.txt') # mengambil semua kata dari pidato pelantikan Roosevelt tahun 1933 (dengan ID file '1933-Roosevelt.txt') dari korpus inaugural NLTK dan menyimpannya dalam variabel bernama inaugural_1933_words. Outputnya adalah daftar kata-kata.
inaugural_1933_words[:20] # menampilkan kata  dari daftar kata yang disimpan dalam variabel inaugural_1933_words.
```

#### 2.c.

```
inaugural_1933_sents = inaugural.sents('1933-Roosevelt.txt') # mengambil semua kalimat dari pidato pelantikan Roosevelt tahun 1933 (dengan ID file '1933-Roosevelt.txt') dari korpus inaugural NLTK dan menyimpannya dalam variabel bernama inaugural_1933_sents. Outputnya adalah daftar kalimat, di mana setiap kalimat adalah daftar kata-kata.
inaugural_1933_sents[:5] #menampilkan 5 kalimat pertama dari daftar kalimat yang disimpan dalam variabel inaugural_1933_sents.
```

#### 2.d.  

```
inaugural_1933_pars = inaugural.paras('1933-Roosevelt.txt') # mengambil semua paragraf dari pidato pelantikan Roosevelt tahun 1933 (dengan ID file '1933-Roosevelt.txt') dari korpus inaugural NLTK dan menyimpannya dalam variabel bernama inaugural_1933_pars. Outputnya adalah daftar paragraf, di mana setiap paragraf adalah daftar kalimat, dan setiap kalimat adalah daftar kata-kata. Ini adalah struktur data yang lebih bersarang.
inaugural_1933_pars[:5] #  menampilkan 5 paragraf pertama dari daftar paragraf yang disimpan dalam variabel inaugural_1933_pars.
```
